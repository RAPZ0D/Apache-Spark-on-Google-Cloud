## This Assignment makes use of the Pubsub Lite which a messaging system on GCP

What is this program doing is that it takes in the data from my bucket and then the python file will take in the data and send the messages to pubsub Lite one by one 

Importing the Library

```python
import json
from google.cloud import storage
from google.cloud.pubsublite.cloudpubsub import PublisherClient
from google.cloud.pubsublite.types import MessageMetadata
import datetime
```

```python
def publish_to_pubsublite():
    # Set your bucket URI and topic
    bucket_uri = "gs://pysparkbucket-jm"
    topic_name = "projects/617930613220/locations/us-central1/topics/pubsubtest"
    
    # Initialize the storage client and get the bucket
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_uri.replace("gs://", ""))
    
    # Get the blob (JSON file) from the bucket
    blob = bucket.blob("events.json")
    data = json.loads(blob.download_as_string())

    # Initialize a Pub/Sub Lite publisher for publishing messages to the topic
    with PublisherClient() as publisher_client:
        for row in data:
            now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            row['event_date'] = now
            message_data = json.dumps(row).encode("utf-8")

            # Publish the message to the Pub/Sub Lite topic
            api_future = publisher_client.publish(topic_name, data=message_data)
            message_id = api_future.result()
            message_metadata = MessageMetadata.decode(message_id)

            print(f"Published message with ID: {message_id} to partition {message_metadata.partition.value} and offset {message_metadata.cursor.offset}.")

    print("Data published to Pub/Sub Lite successfully.")

if __name__ == "__main__":
    publish_to_pubsublite()
```

```python
def publish_to_pubsublite():
    # Set your bucket URI and topic
    bucket_uri = "gs://pysparkbucket-jm"
    topic_name = "projects/617930613220/locations/us-central1/topics/pubsubtest"
    
    # Initialize the storage client and get the bucket
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_uri.replace("gs://", ""))
    
    # Get the blob (JSON file) from the bucket
    blob = bucket.blob("events.json")
    data = json.loads(blob.download_as_string())

    # Initialize a Pub/Sub Lite publisher for publishing messages to the topic
    with PublisherClient() as publisher_client:
        # Counter to track the number of messages sent
        count = 0

        for row in data:
            now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            row['event_date'] = now
            message_data = json.dumps(row).encode("utf-8")

            # Publish the message to the Pub/Sub Lite topic
            api_future = publisher_client.publish(topic_name, data=message_data)
            message_id = api_future.result()
            message_metadata = MessageMetadata.decode(message_id)

            print(f"Published message with ID: {message_id} to partition {message_metadata.partition.value} and offset {message_metadata.cursor.offset}.")
            print("Message Data:", row)  # Print the data being sent along with the message

            # Increment the counter
            count += 1

            # Check if the limit of 10 messages has been reached
            if count >= 10:
                break

    print("Data published to Pub/Sub Lite successfully.")

if __name__ == "__main__":
    publish_to_pubsublite()
```
